Базові стратегії відбору за
невизначеністю

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import Sequential
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from keras.callbacks import EarlyStopping
import time
```

```python
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
initial_size = 1000
X_train, X_pool, y_train, y_pool = train_test_split(X_train, y_train, train_size=initial_size, stratify=y_train)
```

```python
# Створення моделі
def create_model():
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        Conv2D(64, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Dropout(0.25),
        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(10, activation='softmax')  
    ])
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Методи відбору
def least_confidence(predictions):
    return 1 - np.max(predictions, axis=1)

def margin_sampling(predictions):
    sorted_preds = -np.sort(-predictions, axis=1)
    return sorted_preds[:, 0] - sorted_preds[:, 1]

def ratio_confidence(predictions):
    sorted_preds = -np.sort(-predictions, axis=1)
    return sorted_preds[:, 1] / sorted_preds[:, 0]

def entropy(predictions):
    return -np.sum(predictions * np.log(predictions + 1e-10), axis=1)
```

```python
def active_learning(model, X_train, y_train, X_pool, y_pool, method, budget=1000):
    results = []
    for iteration in range(5):
        predictions = model.predict(X_pool)
        if method == 'least_confidence':
            uncertainty = least_confidence(predictions)
        elif method == 'margin_sampling':
            uncertainty = margin_sampling(predictions)
        elif method == 'ratio_confidence':
            uncertainty = ratio_confidence(predictions)
        elif method == 'entropy':
            uncertainty = entropy(predictions)
        
        # Вибір найбільш невизначених зразків
        indices = np.argsort(uncertainty)[:budget]
        X_selected, y_selected = X_pool[indices], y_pool[indices]

        # Додавання обраних зразків до тренувальної вибірки
        X_train = np.concatenate([X_train, X_selected])
        y_train = np.concatenate([y_train, y_selected])

        # Видалення обраних зразків з пулу
        X_pool = np.delete(X_pool, indices, axis=0)
        y_pool = np.delete(y_pool, indices, axis=0)

        early_stopping = EarlyStopping(monitor='val_loss', patience=3)

        # Навчання моделі на новій вибірці
        start_time = time.time()
        model.fit(X_train, y_train, 
                  epochs=5, 
                  batch_size=64, 
                  validation_data=(X_test, y_test), 
                  verbose=1, 
                  callbacks=[early_stopping])
        end_time = time.time()

        # Оцінка моделі
        _, train_acc = model.evaluate(X_train, y_train, verbose=0)
        _, test_acc = model.evaluate(X_test, y_test, verbose=0)
        results.append({
            'iteration': iteration,
            'train_accuracy': train_acc,
            'test_accuracy': test_acc,
            'execution_time': end_time - start_time,
            'selected_samples': len(X_selected)
        })

    return results
```

```python
# Основний цикл
methods = ['least_confidence', 'margin_sampling', 'ratio_confidence', 'entropy']
results_table = []

for method in methods:
    print(f"Running with method: {method}")
    model = create_model()

    early_stopping = EarlyStopping(monitor='val_loss', patience=3)
    
    model.fit(X_train, y_train, 
              epochs=5, 
              batch_size=64, 
              validation_data=(X_test, y_test), 
              verbose=1, 
              callbacks=[early_stopping]) 
    
    results = active_learning(model, X_train, y_train, X_pool, y_pool, method)

    for result in results:
        results_table.append({
            'method': method,
            'iteration': result['iteration'] + 1,
            'train_accuracy': result['train_accuracy'],
            'test_accuracy': result['test_accuracy'],
            'execution_time': result['execution_time'],
            'selected_samples': result['selected_samples']
        })
```

```python
# Виведення результатів
import pandas as pd
df = pd.DataFrame(results_table)
print(df)
```

```python
import seaborn as sns

iterations = [1, 2, 3, 4, 5]

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.lineplot(data=df, x='iteration', y='train_accuracy', hue='method', marker="o")
plt.title('Навчальна точність по ітераціях')
plt.xlabel('Ітерація')
plt.ylabel('Навчальна точність')
plt.xticks(ticks=iterations)

plt.subplot(1, 2, 2)
sns.lineplot(data=df, x='iteration', y='test_accuracy', hue='method', marker="o")
plt.title('Тестова точність по ітераціях')
plt.xlabel('Ітерація')
plt.ylabel('Тестова точність')
plt.xticks(ticks=iterations)

plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='iteration', y='execution_time', hue='method', marker="o")
plt.title('Час виконання по ітераціях')
plt.xlabel('Ітерація')
plt.ylabel('Час виконання (секунди)')
plt.xticks(ticks=iterations)
plt.show()
```
