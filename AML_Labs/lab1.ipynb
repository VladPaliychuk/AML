```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
x_train_copy = np.copy(x_train)
y_train_copy = np.copy(y_train)
```

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, 
          epochs=10, 
          batch_size=64,
          validation_split=0.2,
          verbose=1)

# Оцінка моделі на тестовому наборі
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print('\nTest accuracy:', test_acc)

# Отримання прогнозів для тестового набору
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Розрахунок метрик
print('Accuracy:', accuracy_score(y_test, y_pred_classes))
print('Precision:', precision_score(y_test, y_pred_classes, average='macro'))
print('Recall:', recall_score(y_test, y_pred_classes, average='macro'))
print('F1-score:', f1_score(y_test, y_pred_classes, average='macro'))
```

```python
# Розподіл на навчальну, тестову та пул нерозмічених даних
initial_size = 1000
X_train = x_train_copy[:initial_size]
y_train = y_train_copy[:initial_size]
X_pool = x_train_copy[initial_size:]
y_pool = y_train_copy[initial_size:]

#Функція для розрахунку невизначеності (ентропія прогнозів)
def get_uncertainty(model, X):
    probs = model.predict(X)
    entropy = -np.sum(probs * np.log2(probs), axis=1)
    return entropy

model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
#model.add(Dropout(0.5))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Навчання початкової моделі
model.fit(X_train, y_train, 
          epochs=10, 
          batch_size=64,
          validation_data=(x_test, y_test),
          verbose=1)

train_losses = []
val_losses = []
train_accuracies = []
val_accuracies = []

# Ітерації активного навчання
for iteration in range(5):
    # Розрахунок невизначеності для пулу даних
    uncertainty = get_uncertainty(model, X_pool)

    # Вибір найбільш невизначених зображень
    budget = int(1000 * (0.8 ** iteration))  # Зменшуємо бюджет на 20% на кожній ітерації
    indices = np.argsort(uncertainty)[-budget:]
    X_selected, y_selected = X_pool[indices], y_pool[indices]

    # Додавання нових даних до навчальної вибірки
    X_train = np.concatenate([X_train, X_selected])
    y_train = np.concatenate([y_train, y_selected])

    # Видалення обраних даних з пулу
    X_pool = np.delete(X_pool, indices, axis=0)
    y_pool = np.delete(y_pool, indices, axis=0)

    # Навчання моделі на оновленому наборі даних
    history = model.fit(X_train, y_train, 
                        epochs=5, 
                        batch_size=128,
                        validation_data=(x_test, y_test), 
                        verbose=1)

    train_losses.extend(history.history['loss'])
    val_losses.extend(history.history['val_loss'])
    train_accuracies.extend(history.history['accuracy'])
    val_accuracies.extend(history.history['val_accuracy'])
    
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
    print('\nTest accuracy:', test_acc)
```

```python
# Отримання прогнозів для тестового набору
y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Розрахунок метрик
print('Accuracy:', accuracy_score(y_test, y_pred_classes))
print('Precision:', precision_score(y_test, y_pred_classes, average='macro'))
print('Recall:', recall_score(y_test, y_pred_classes, average='macro'))
print('F1-score:', f1_score(y_test, y_pred_classes, average='macro'))
```

```python
# Графік для втрат
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title('Втрати (Loss)')
plt.xlabel('Епохи/Ітерації')
plt.ylabel('Loss')
plt.legend()

# Графік для точності
plt.subplot(1, 2, 2)
plt.plot(train_accuracies, label='Train Accuracy')
plt.plot(val_accuracies, label='Validation Accuracy')
plt.title('Точність (Accuracy)')
plt.xlabel('Епохи/Ітерації')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
